```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('HR Employee Attrition.csv')
```

**Coverting Attrition into Numeric**

```python
label_encode = LabelEncoder()

labels = label_encode.fit_transform(data['Attrition'])

data['Attrition'] = labels
```

```python
data['Attrition'].value_counts()
```

```python
no=data[data.Attrition==0]
yes=data[data.Attrition==1]
```

```python
print(no)
```

**Under Sampling**

```python
no_sample=no.sample(n=237)
```

```python
new_dataset=pd.concat([no_sample,yes],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Attrition'].value_counts()
```

**Removing Unnecessary Data**

```python
new_dataset.drop(columns=['StandardHours','EmployeeCount','EducationField','MaritalStatus','Over18','MonthlyRate'],axis=1)
```

**Coverting Categorical data into Numeric**

```python
numeric_features=[]
categorical_features=[]
for col in new_dataset.columns:
    if new_dataset[col].dtype=='int64':
        numeric_features.append(col)
    else:
        categorical_features.append(col)
categorical_features
```

```python
Y=new_dataset["Attrition"]
Y
```

```python
X=new_dataset.drop(columns=['Attrition'])
mappings=[]

for features in categorical_features:
  mapping={}
  temp=0

  for key in set(X[features]):
    mapping[key]=temp
    temp+=1
  X[features]= X[features].map(mapping)
  mappings.append(mapping)
  
for i in range(len(categorical_features)):
  print(categorical_features[i])
  print(mappings[i])
```

**Splitting Of Data**

```python
X_new, X_test, y_new, y_test = train_test_split(X, Y, test_size=0.1, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_new, y_new, test_size=0.11, random_state=0)
```

```python
print(X_train.shape) #Training Data
print(y_train.shape)
print(X_test.shape) #Testing Data
print(y_test.shape)
print(X_val.shape) #Validation Data
print(y_val.shape)
```

**Decision Tree Classification Model**

```python
classifier= DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
```

**Accuracy**

```python
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
print('Accuracy: ', training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,y_test)
print('Accuracy: ', test_data_accuracy)
```
